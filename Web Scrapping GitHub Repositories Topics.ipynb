# Use Beautiful Soup to parse and extract information

```python
#installing and importing of Beautiful soap library
!pip install beautifulsoup4 --upgrade --quiet
from bs4 import BeautifulSoup
```

```python
 #1st step we extraxt topic title from html
doc=BeautifulSoup(page_content,'html.parser')
selection_class = 'f3 lh-condensed mb-0 mt-1 Link--primary'

topic_title_tags = doc.find_all('p', {'class': selection_class})
```

```python
topic_title_tags[4].text.strip()        
```

```python
#2nd step we extract topoc discription from html
desc_selector = 'f5 color-fg-muted mb-0 mt-1'

topic_desc_tags = doc.find_all('p', {'class': desc_selector})
topic_desc_tags[4].text.strip()                                
```

```python
topic_title_tag0 = topic_title_tags[0]
```

```python
div_tag = topic_title_tag0.parent
```

```python
topic_link_tags=doc.find_all('a',class_="no-underline flex-grow-0")
```

```python
#we extract topic url by adding "https://github.com" from html
topic0_url = "https://github.com" + topic_link_tags[0]['href']    
print(topic0_url)
```

```python
#4th step we extract the list of titles by looping
topic_titles=[]
for tag in topic_title_tags:
    topic_titles.append(tag.text)      
        
topic_titles
len(topic_titles)
```

```python
 #5th step we extract the list of discription by looping 
topic_descs = []

for desc in topic_desc_tags:
    topic_descs.append(desc.text.strip())    
    
topic_descs
len(topic_descs)
```

```python
 #6th step we extract the list of by looping 
topic_urls = []
base_url = 'https://github.com'


for tag in topic_link_tags:
    topic_urls.append(base_url + tag['href'])    
    
topic_urls
len(topic_urls)
```

```python
topics_dict = {
    'title': topic_titles,
    'description': topic_descs, #we convert those individual data(list) into dict format to add them into dataframe directly
    'url': topic_urls
}
```

```python
import pandas as pd

topics_df = pd.DataFrame.from_dict(topics_dict)
topics_df.reset_index(drop=True, inplace=True)
topics_df
```

```python
star_tags = doc.find_all(id="repo-stars-counter-star", class_= 'Counter js-social-count') #to find the star class
star_tags[0].text.strip()
```

```python
def parse_star_count(stars_str):
    stars_str = stars_str.strip()
    if stars_str[-1] == 'k':
        return int(float(stars_str[:-1]) * 1000)  #function to convert float format into int
    return int(stars_str)
```

```python
parse_star_count(star_tags[0].text.strip())
```

```python
def get_repo_info(h3_tag, star_tag):
    # returns all the required info about a repository
    a_tags = h3_tag.find_all('a')
    username = a_tags[0].text.strip()
    repo_name = a_tags[1].text.strip()
    repo_url =  base_url + a_tags[1]['href']      #function to extract username,repo-name,star,URL
    stars = parse_star_count(star_tag.text.strip())
    return username, repo_name, stars, repo_url
```

```python
get_repo_info(ref_tag[0], star_tags[0])
```

```python
import pandas as pd
#create dict to add the Extracted data into dictionary
topic_repos_dict = {
    'username': [],
    'repo_name': [],
    'stars': [],
    'repo_url': []
}
for i in range(len(ref_tag)):
    repo_info = get_repo_info(ref_tag[i], star_tags[i])
    topic_repos_dict['username'].append(repo_info[0])
    topic_repos_dict['repo_name'].append(repo_info[1])
    topic_repos_dict['stars'].append(repo_info[2])
    topic_repos_dict['repo_url'].append(repo_info[3])
  
topic_repos_dict  
#Convert dict into DataFrame
file=pd.DataFrame(topic_repos_dict)
file
```
